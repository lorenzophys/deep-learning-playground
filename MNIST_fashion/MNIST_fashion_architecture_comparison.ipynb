# MNIST fashion architecture comparison.

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
ensemble_DNN = {"mu": 0.8934, "std": 0.0018}
ensemble_CNN = {"mu": 0.9340, "std": 0.0014}
```

```python
x = np.arange(0.880, 0.96, 0.00001)
y_DNN =  norm(ensemble_DNN["mu"], ensemble_DNN["std"])
y_CNN = norm(ensemble_CNN["mu"], ensemble_CNN["std"])

plt.figure(figsize=(15,5))

plt.plot(x, y_DNN.pdf(x))
plt.plot(x, y_CNN.pdf(x))

plt.xlabel('accuracy')
plt.legend(["DNN", "CNN"])

plt.savefig('../img/MNIST_fashion_comparison.png')

plt.show()
```
